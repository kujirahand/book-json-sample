```python
import pandas as pd
df = pd.read_csv("kamakura.csv", encoding='utf-8')
df
```

```python
df.columns
```

```python
# 「宅地(土地と建物)」を抽出
df2 = df[df['種類'] == '宅地(土地と建物)']
df2
```

```python
# 必要な項目だけを抽出
df3 = df2[['取引価格（総額）',  '面積（㎡）', '間口', '延床面積（㎡）', '地区名', '最寄駅：名称', '最寄駅：距離（分）', '前面道路：幅員（ｍ）', '都市計画', '建ぺい率（％）', '容積率（％）', '建物の構造']].copy()
df3
```

```python
from sklearn.preprocessing import LabelEncoder
station_enc = LabelEncoder()
df3['最寄駅：名称'] = station_enc.fit_transform(df3['最寄駅：名称'])
area_enc = LabelEncoder()
df3['地区名'] = area_enc.fit_transform(df3['地区名'])
toshi_enc = LabelEncoder()
df3['都市計画'] = toshi_enc.fit_transform(df3['都市計画'])
kozou_enc = LabelEncoder()
df3['建物の構造'] = kozou_enc.fit_transform(df3['建物の構造'])
```

```python
# 数値に変換できない値を削除
df3['最寄駅：距離（分）'] = pd.to_numeric(df3['最寄駅：距離（分）'], errors='coerce')
df3['面積（㎡）'] = pd.to_numeric(df3['面積（㎡）'], errors='coerce')
df3['間口'] = pd.to_numeric(df3['間口'], errors='coerce')
df3['面積（㎡）'] = pd.to_numeric(df3['面積（㎡）'], errors = 'coerce')
df3['延床面積（㎡）'] = pd.to_numeric(df3['延床面積（㎡）'], errors = 'coerce')
df3  = df3.dropna()
df3[df3['延床面積（㎡）'].isnull()]
```

```python
y = df3['取引価格（総額）']
x = df3.drop(columns=['取引価格（総額）'])
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

x_train, x_test, y_train, y_test = train_test_split(x, y)

model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

train_df_corr = df3.corr()
sns.heatmap(train_df_corr, vmax=1, vmin=-1, center=0, annot=True)
plt.tight_layout()
plt.show()

```
