```python
from sklearn import svm
import random

text = open('agaricus-lepiota.data', 'r').read()

lines = text.split('\n')
random.shuffle(lines)

x, y = [], []
for line in text.split('\n'):
    if line == '': continue
    cells = line.split(',')
    y.append(cells[0])
    x.append(cells[1:])
```

```python
def label_to_num(rows, index):
    # どんなラベルがあるか調べて番号に置き換える
    labels = {}
    id = 0
    for row in rows:
        v = row[index]
        if v not in labels:
            labels[v] = id
            id += 1
        row[index] = labels[v]

for i in range(22):
    label_to_num(x, i)
```

```python
# 学習データとテストデータに分ける
n = int(len(lines) * 0.8)
x_train = x[0:n]
y_train = y[0:n]
x_test = x[n:]
y_test = y[n:]

clf = svm.SVC()
clf.fit(x_train, y_train)

pre = clf.predict(x_test)

ok = 0
for i, v in enumerate(pre):
    if v == y_test[i]: ok += 1
print('ok=,', ok, 'total=', len(y_test))
print('rate=', ok / len(y_test))
```
