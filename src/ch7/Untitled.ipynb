```python
import pandas as pd
df = pd.read_csv('yokohama-kanagawaku.csv', encoding='cp932')
```

```python
df2 = df[df['種類'] == '中古マンション等']
df2
```

```python
# 必要な項目だけを抽出
# df3 = df2[['取引価格（総額）',  '面積（㎡）', '間口', '地区名', '最寄駅：名称', '最寄駅：距離（分）', '前面道路：幅員（ｍ）', '都市計画', '建ぺい率（％）', '容積率（％）', '建物の構造']].copy()
df3 = df2[['取引価格（総額）',  '面積（㎡）', '地区名', '最寄駅：距離（分）', '都市計画', '建物の構造', '建築年', '取引時点']].copy()
df3
```

```python
# 最寄駅までの距離データを確認
df3 = df3.replace({'最寄駅：距離（分）': {"1H?1H30": 60}})
df3 = df3.replace({'最寄駅：距離（分）': {"1H30?2H": 90}})
df3 = df3.replace({'最寄駅：距離（分）': {"2H?": 120}})
df3 = df3.replace({'最寄駅：距離（分）': {"30分?60分": 30}})
#df3.groupby('最寄駅：距離（分）').count()
import re
# 各年号の元年を定義
eraDict = {
    "明治": 1868,
    "大正": 1912,
    "昭和": 1926,
    "平成": 1989,
    "令和": 2019,
}
def wareki2seireki(text):
    if type(text) is int: return text
    # 年月日を抽出
    pattern = r"(?P<era>{eraList})(?P<year>[0-9]{{1,2}}|元)年".format(eraList="|".join(eraDict.keys()))
    date = re.search(pattern, str(text))
    # 抽出できなかったら終わり
    if date is None:
        # print("変換できない値", text)
        return 0 # 適当な値を返す
    # 年を変換
    for era, startYear in eraDict.items():
        if date.group("era") == era:
            if date.group("year") == "元":
                year = eraDict[era]
            else:
                year = eraDict[era] + int(date.group("year")) - 1
    return year
# 築年数を計算
df3['建築年'] = df3['建築年'].map(lambda v: wareki2seireki(v))
df3['取引時点'] = df3['取引時点'].str[0:4].astype('int')
df3['築年数'] = df3['取引時点'] - df3['建築年']
df3 = df3.drop(columns=['建築年', '取引時点'])
# 200年以上の物件を除外
df3 = df3[df3['築年数'] < 200]
```

```python
# 一度ファイルに保存
df3.to_csv('yokohama-kanagawaku-m.csv')
```
